```python
#sorting by age of house

housing.sort_values("age")
```

```python
#sorting by age of house descending

housing.sort_values("age", ascending = False)
```

```python
#sorting with multiple columns (with one ascending and other descending)

housing.sort_values(["age","rm"] , ascending = [False,True])
```

```python
#find houses aged more than 50 years age

housing[housing["age"]>50]
```

```python
# min age

housing["age"].min()
```

```python
'''aggregate function using a function to give the inter quartile range. i.e.  
75th percentile minus the 25th percentile this is usefull to detrmine outliers'''
import numpy as np
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)

housing["age"].agg(iqr)
```

```python
'''Multiple column'''
housing[["age","rm"]].agg(iqr)
```

```python
'''use of agg with more than one function'''

print(housing[["age", "rm"]].agg([iqr , np.median]))
```
