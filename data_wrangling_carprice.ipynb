
### Handling Missing Values


Note: None of the columns or rows are empty enough to be dropped completely. 

Decision on Missing Values: 

Replace by mean:

"normalized-losses": 41 missing data, replace them with mean
"stroke": 4 missing data, replace them with mean
"bore": 4 missing data, replace them with mean
"horsepower": 2 missing data, replace them with mean
"peak-rpm": 2 missing data, replace them with mean

Replace by frequency:

"num-of-doors": 2 missing data, replace them with "four".
Reason: More number (84%) is with 4 doors. We assume since it occurs more frequently, it is most likely to occur


Drop the whole row:

"price": 4 missing data, simply delete the whole row
Reason: price is what we want to predict. Any data entry without price data cannot be used for prediction; so row with missing price data can be removed. 

```python
#Replace with mean along axis =0 i.e. all the ROWS in each COLUMN

df["normalized-losses"].replace(np.nan, df["normalized-losses"].astype("float").mean(axis=0), inplace=True)

df["stroke"].replace(np.nan, df["stroke"].astype("float").mean(axis = 0), inplace = True)

df["bore"].replace(np.nan, df ["bore"].astype("float").mean(axis = 0), inplace = True )

df["horsepower"].replace(np.nan, df ["horsepower"].astype("float").mean(axis = 0), inplace = True )

df["peak-rpm"].replace(np.nan, df ["peak-rpm"].astype("float").mean(axis = 0), inplace = True )
```

```python
#replace the missing 'num-of-doors' values by the most frequent 

df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
# drop rows with no value for price and reset index

df.dropna(subset=["price"], axis=0, inplace=True)
df.reset_index(drop = True)
df.head(5)
```

```python
# fix the data types

df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

### Data Standardization

```python
#add mileage in liters

df['city-L/100km'] = 235/df["city-mpg"]

# check transformed data 
df.head()
```

```python
df['highway-L/100'] = 235/df["highway-mpg"]
df.head(5)
```

### Data Normalization   
 to bring variables to similar range.

```python
#simple feature scaling

df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()

df[["length","width","height"]].head()
```

### Binning  
to convert data into discrete categorical bins. 

```python
df["horsepower"]=df["horsepower"].astype(int, copy=True)
plt.pyplot.hist(df["horsepower"])
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# create bin array using linspace
# set group names

bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)

group_names = ['Low', 'Medium', 'High']

df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(5)
```

```python
#count on each bin
df["horsepower-binned"].value_counts()
```

```python
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

## Encoding

```python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
#add the new columns 
df = pd.concat([df, dummy_variable_1], axis =1)

df.head()
```

```python
# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)
```

```python
df.head()
```

```python
# One - hot Encoding for aspiration

dummy_variable_2 = pd.get_dummies(df['aspiration'])

dummy_variable_2.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)

dummy_variable_2.head()
```

```python
df= pd.concat([df, dummy_variable_2], axis =1)
df.drop("aspiration", axis =1, inplace = True)
df.head()
```
